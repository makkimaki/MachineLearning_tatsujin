```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor

x = 10 * np.random.rand(30)

def cos_model(x, sigma=0.3):
    noise = sigma * np.random.randn(len(x))

    return np.sin(1.5*x) + np.cos(0.1 * x) + noise

y = cos_model(x)

plt.figure(figsize=(10,7))
plt.errorbar(x, y, 0.1, fmt="bo")
plt.show()
```

```python
X = np.linspace(0, 10, 1000)

trees = [3, 50, 100]

ytrue = cos_model(X, 0)

plt.figure(figsize=(10,7))
plt.errorbar(x, y, 0.1, fmt="bo")
plt.plot(X, ytrue, "-k", alpha=0.4, label="True")

for tree in trees:
    rfr = RandomForestRegressor(tree)
    rfr.fit(x[:, None], y)
    yfit = rfr.predict(X[:, None])

    plt.plot(X, yfit, "-", label=tree)

plt.legend()
```
